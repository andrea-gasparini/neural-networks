```python
import matplotlib.pyplot as plt
from data_generator import DataGenerator
import numpy as np
import matplotlib
```

```python
matplotlib.use("pgf")
matplotlib.rcParams.update({
    "pgf.texsystem": "pdflatex",
    'font.family': 'serif',
    'text.usetex': True,
    'pgf.rcfonts': False,
})
```

```python
ax, fig = plt.subplots(figsize=(4, 4))

generator = DataGenerator(2)
values, _ = generator.generate(500, 0)
values2, _ = generator.generate(500, 2)

fig.scatter(values[:,0], values[:,1], label="mean = 0")
fig.scatter(values2[:,0], values2[:,1], color="red", label="mean = 2")
fig.set_xticks(range(-4, 7, 2))
fig.set_yticks(range(-4, 7, 2))
fig.set_xlabel("Dimension 1")
fig.set_ylabel("Dimension 2")
fig.legend()

plt.savefig("pictures/data-generation-2.pgf")
```
