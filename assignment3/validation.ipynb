```python
import numpy as np
import matplotlib.pyplot as plt
from src.network import Network
from src.data import Data
```

```python
X = np.transpose(np.genfromtxt("xi.csv", delimiter=","))
Y = np.transpose(np.genfromtxt("tau.csv", delimiter=","))
N = 50
```

```python
K_range = [2, 5, 8, 12, 20]

tr_e = []
ts_e = []

for K in K_range:
    print(f"Doing {K = }")

    network = Network(N, K)

    train, test = Data.sample(X, Y)
    train_errors, test_errors = network.train(train, test, t_max=100)

    tr_e.append(train_errors)
    ts_e.append(test_errors)
```

```python
fig, plots = plt.subplots(1, len(K_range), figsize=(20, 4))

for plot, K, test_error, train_error in zip(plots, K_range, ts_e, tr_e):
    plot.plot(test_error, label="test")
    plot.plot(train_error, label="train")
    plot.legend()
```

```python
fig, plots = plt.subplots(1, len(K_range), figsize=(20, 4))

for plot, K, test_error, train_error in zip(plots, K_range, ts_e, tr_e):
    plot.plot(train_error[20:], label="train")
    plot.legend()
    plot.set_title(f"{K = }")
```
